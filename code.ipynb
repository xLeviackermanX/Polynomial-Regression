```python
import pickle
from prettytable import PrettyTable
import math
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import random
import numpy as np
from datetime import datetime
random.seed(datetime.now())


filename = "./data/train.pkl"
infile = open(filename,'rb')
arr = pickle.load(infile)
infile.close()
#print(type(arr))
filename = "./data/test.pkl"
infile = open(filename,'rb')
test = pickle.load(infile)
infile.close()
#print(arr)
k =int( len(arr))
n =int( k/10)
ind = []
for i in range(0,k):
    ind.append(i);

random.shuffle(ind)
x = []
y = []
xt= []
yt= []
yt1 = []
bias2 = []
bias = []
variance = []
complexity = []
for i in range(1,21):
    row = []
    row.append(i)
    complexity.append(row)
error = []
#print(x)
#print(int(k/10))
for i in range(0,10):
    row1 = []
    row2 = []
    for j in range(0,n):
        cols1 = []
        cols1.append(arr[i*n+j][0])
        row1.append(cols1)
        cols2 =[]
        cols2.append(arr[i*n+j][1])
        row2.append(cols2)
    x.append(row1)
    y.append(row2)
#print(x[0])
k = len(test)
for i in range(0,k):
    row1 = []
    row2 = []
    row1.append(test[i][0])
    xt.append(row1)
    row2.append(test[i][1])
    yt.append(row2)

yt = np.array(yt)
table = PrettyTable() 
table.field_names = ["Degree of Polynomial","Bias","Variance","Irreducible Error"]

for deg in range(1,21):
    curBias = 0.0
    curBias2 = 0.0
    curVariance= 0.0
    ourY = []
    ourY2 = []
    ourErr = []
    curTotalErr = 0.0
    for j in range(0,10):
        poly_reg = PolynomialFeatures(degree=deg)
        x1= poly_reg.fit_transform(x[j])
        reg = LinearRegression()
        reg.fit(x1,y[j])
        y1 = reg.predict(poly_reg.fit_transform(xt))
        y2 = abs(y1-yt)
        y2 = np.square(y2)
        ourErr.append(y2)
        ourY.append(y1)
        y1 = np.square(y1)
        ourY2.append(y1)
    s3 = np.sum(ourErr, axis=0)
    s3 = s3/10
    s = np.sum(ourY, axis=0)
    s=np.array(s)
    s = s/10
    s2 = np.sum(ourY2 , axis = 0)
    s2 = np.array(s2)
    s2 = s2/10
    s1 = np.square(s)
    res= s2-s1
    curVariance = np.sum(res)
    curVariance/=k
    res2 = s-yt
    res2 = np.square(res2)
    curBias2 = np.sum(res2)
    curBias2/=k
    res3 = abs(s-yt)
    curBias = np.sum(res3)
    curTotalErr = np.sum(s3)
    curTotalErr/=k
    curBias=curBias/k
    row = []
    row.append(curBias2)
    bias2.append(row)
    row = []
    row.append(curVariance)
    variance.append(row)
    row = []
    row.append(curTotalErr)
    error.append(row)
    table.add_row([deg ,curBias,curVariance,curTotalErr-curBias2-curVariance])
filename = 'table.txt'
f = open(filename,'w')
print(table,file=f)
plt.plot(complexity,bias2,color='red',label='Bias^2')
plt.plot(complexity,variance, color='blue',label='Variance')
plt.plot(complexity,error, color='green',label='MSE')
plt.ylabel('Error')
plt.xlabel('Complexity')
plt.legend()
plt.show()
```
